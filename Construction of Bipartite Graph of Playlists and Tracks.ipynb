```python
from stellargraph import StellarGraph
import pandas as pd
```

```python
edges = []
tracks = []
playlists = []
for i in range(0, 15000, 1000):
    edges.append(pd.read_csv(f'D:\\million_playlist_dataset_csv\\playlist_tracks_{i}-{i+999}.csv'))
    tracks.append(pd.read_csv(f'D:\\million_playlist_dataset_csv\\tracks_{i}-{i+999}.csv'))
    playlists.append(pd.read_csv(f'D:\\million_playlist_dataset_csv\\playlists_{i}-{i+999}.csv'))
edges = pd.concat(edges, ignore_index=True)
#edges = edges.drop('pos',axis=1)
#edges.columns = ['target', 'source']
edges
```

```python
import csv

with open('D:\\extracted_features\\all_edges.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    
    writer.writerow(edges.columns)
    
    writer.writerows(edges.values)
```

```python
playlists = pd.concat(playlists)
playlists = playlists.drop(['name', 'collaborative'], axis=1)
playlists.index=playlists['pid']
playlists
```

```python
tracks = pd.concat(tracks)
tracks = pd.DataFrame(tracks.groupby(tracks['track_uri']))
tracks[1]=0
tracks
tracks.index=tracks[0]
tracks[0]=0
tracks
```

```python
playlists = StellarGraph({'playlist': playlists, 'track': tracks} ,edges)
print(playlists.info())
```
