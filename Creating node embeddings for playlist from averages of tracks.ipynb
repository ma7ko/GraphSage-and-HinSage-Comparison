```python
import pandas as pd

tracks = pd.read_csv('D:\\extracted_features\\tracks_256_full.csv')
tracks.head()
```

```python
playlists_tracks = pd.read_csv(f'D:\\extracted_features\\train_edges_graph.csv')
playlists_tracks
```

```python
playlists_tracks = playlists_tracks.drop('pos', axis=1)
playlists_tracks
```

```python
playlist_embeddings = []
for i in range (0, 15000):
    current_playlist = playlists_tracks[playlists_tracks['pid'] == i]
    #print(current_playlist.set_index('track_uri').join(tracks.set_index('track_uri'), how='inner'))
    current_playlist = current_playlist.set_index('track_uri').join(tracks.set_index('track_uri'), how='inner')
    playlist_vector = [current_playlist[f"{j}"].mean() for j in range(0, 256)]
    playlist_vector.insert(0,i)
    playlist_embeddings.append(playlist_vector)
```

```python
import csv

csv_headers = ['pid']
[csv_headers.append(i) for i in range(0, 256)]

with open('D:\\extracted_features\\playlists_256_full.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    
    writer.writerow(csv_headers)
    
    writer.writerows(playlist_embeddings)
```
