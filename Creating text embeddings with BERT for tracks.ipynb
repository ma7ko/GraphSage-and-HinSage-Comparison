```python
import pandas as pd

tracks = []
for i in range(0, 15000, 1000):
    tracks.append(pd.read_csv(f'D:\\million_playlist_dataset_csv\\tracks_{i}-{i+999}.csv',  dtype=str))
tracks = pd.concat(tracks, ignore_index=True)
tracks
```

```python
tracks.index = tracks['track_uri']
tracks
```

```python
#tracks = tracks.groupby(tracks['track_uri'], axis=0)
grouped_tracks = pd.DataFrame(tracks.first())
grouped_tracks['track_uri'] = grouped_tracks.index
grouped_tracks
```

```python
import csv

with open('D:\\extracted_features\\all_tracks.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    
    writer.writerow(grouped_tracks.columns)
    
    writer.writerows(grouped_tracks.values)
```

```python
for uri in ['artist_uri', 'album_uri']:
    grouped_tracks[uri] = [value.split(':')[2] for value in grouped_tracks[uri].values]
    
grouped_tracks['track_uri'] =  [value.split(':')[2] for value in grouped_tracks.index]
grouped_tracks
```

```python
tracks_joined = [' '.join([str(item).lower() for item in row]) for row in grouped_tracks.values]
print(len(tracks_joined[0]))
tracks_joined[:10]
```

```python
from ipynb.fs.full.bert_helper import create_raw_input, pull_bert, get_max_lengths, bert_encode

bert_layer, tokenizer = pull_bert('https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-2_H-256_A-4/2')
X = bert_encode(tracks_joined, tokenizer, 140)
```

```python
X
```

```python
import tensorflow as tf

input_word_ids = tf.keras.layers.Input(shape=(140,), dtype=tf.int32, name="input_word_ids")
input_mask = tf.keras.layers.Input(shape=(140,), dtype=tf.int32, name="input_mask")
input_type_ids = tf.keras.layers.Input(shape=(140,), dtype=tf.int32, name="input_type_ids")

bert_input = {
    'input_word_ids': input_word_ids,
    'input_mask': input_mask,
    'input_type_ids': input_type_ids
}

outputs = bert_layer(bert_input)
pooled_output = outputs['pooled_output']
sequence_output = outputs['sequence_output']
model = tf.keras.Model(
    inputs={
        'input_word_ids': input_word_ids,
        'input_mask': input_mask,
        'input_type_ids': input_type_ids
    },
    outputs=pooled_output
)

model.summary()
```

```python
extracted_features = model.predict(X)
extracted_features
```
